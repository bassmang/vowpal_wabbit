### Class to Simulate Geometric Model Scopes without Champ Changes

```python
import matplotlib.pyplot as plt
import copy

class dec_exp:
    def __init__(self, n = 6, examples = 1000000):
        self.n = n
        self.examples = examples
        self.freqs = []
        self.size_arrs = []
        
    def simulate(self):
        self.freqs = [0]*self.n
        sizes = [0]*self.n
        self.size_arrs = []
        for ex_ind in range(1,self.examples):
            for size_ind in reversed(range(len(sizes))):
                sizes[size_ind] += 1
                if sizes[size_ind] > ex_ind**((size_ind + 1) / self.n):
                    for new_size in reversed(range(size_ind)):
                        sizes[new_size + 1] = sizes[new_size]
                    sizes[0] = 0
                    self.freqs[size_ind] += 1
            self.size_arrs.append(copy.copy(sizes))
    
    def plot_change_freqs(self):
        plt.plot([f / self.examples for f in self.freqs[:len(self.freqs)-1]])
        plt.yscale("log")
        plt.title("Frequency of Resets (Log Scale)", fontsize=18)
        plt.xlabel("Model Index", fontsize=14)
        plt.ylabel("Reset Frequency", fontsize=14)
        plt.show()
        
    def plot_model_scopes(self):
        for i in range(self.n):
            plt.title("Model " + str(i) + " Scope Through Examples", fontsize=18)
            plt.xlabel("Total Examples", fontsize=14)
            plt.ylabel("Examples In Scope", fontsize=14)
            plt.plot([s[i] for s in self.size_arrs])
            plt.show()
```

### Frequency Each Model Passes Geometric Threshold

```python
d = dec_exp(n = 6, examples = 1000000)
d.simulate()
d.plot_change_freqs()
```

### Scope of Each Model Throughout Examples

```python
d.plot_model_scopes()
```
